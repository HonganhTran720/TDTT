## W2A4 : [AverageOfSix] 
Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng 
6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó: 
• Các điểm a1, b1, c1 được nhân hệ số 1, 
• Các điểm a2, b2 được nhân hệ số 2, 
• Điểm a3 được nhân hệ số 3. 
Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10 
Đầu vào: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6 
đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được 
phân tách nhau bởi duy nhất 1 dấu cách. 
Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân. 

```python
a1, b1, c1, a2, b2, a3 = map(int, input().split())
TB = ((a1+b1+c1) + (a2+b2)*2 + a3*3)/10
print(f"{TB:.1f}")
```

## W2A6 [StringDataType] 
“Chuỗi” trong python là các kí tự được xâu lại với nhau như những hạt trên một chiếc dây chuyền. 
Chính vì thế “chuỗi” còn được gọi là “xâu”. Một chuỗi gồm các kí tự nằm trong dấu nháy đơn '' hoặc 
nháy đôi "”.• Mỗi chuỗi (str) là một chuỗi các kí tự trong bảng mã Unicode kể từ Python 3 trở đi 
(Unicode là một tiêu chuẩn mã hóa kí tự toàn cầu, cho phép biểu diễn hầu hết mọi kí tự trong tất cả 
các ngôn ngữ trên thế giới, bao gồm cả kí hiệu, biểu tượng, emoji,….). Bạn có thể dùng ord() và 
chr() để làm việc với mã Unicode.  
Hãy nhập vào một kí tự chữ cái thường ‘a’…’z’, in ra mã Unicode của nó sau đó chuyển nó thành kí 
tự hoa tương ứng, in kí tự hoa này ra màn hình.

```python
u = input()
c = ord(u)
print(f"{chr(c - 32)} {c}")
```

## W2A8 [CelsiusTranfer] 
Fahrenheit, hay độ F, là một thang nhiệt độ được đặt theo tên nhà vật lý người Đức Daniel 
Gabriel Fahrenheit (1686–1736). Dựa vào nhiệt độ Celcius được nhập từ bàn phím, hãy 
chuyển sang độ F theo công thức: F = 9/5*C + 32 và in ra màn hình kết quả. Kết quả làm 
tròn đến hai chữ số thập phân đằng sau dấu phẩy.

```python
C = float(input())
F = 9/5 * C + 32
print(f"{F:.2f}")
```
