## W5A1 
Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> 
int: 

```python
def Max_Of_Two(a, b):
    if (a > b):
        return(a)
    else:
        return(b)

so1, so2 = map(int, input().split())
ketqua = Max_Of_Two(so1, so2)
print(ketqua)
```

## W5A2 Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]: 

```python
def swap(a,b):
    a, b = b, a
    return a, b

x, y = map(int, input().split())
ketqua = swap(x, y)
print(ketqua)
```

## W5A3: Hàm kiểm tra số nguyên tố

```python
def songuyento(n):
    if n < 2:
        return False
    
    for i in range(2, int(n ** 0.5)+1):
        if n % i == 0:
            return False
        else:
            return True

a = int(input())
print(songuyento(a))
```

## W5A4: Hàm kiểm tra số hoàn hảo


```python
def sohoanhao(a):
    if a < 0:
        return False
    
    tong = 0
    
    for i in range(1, a):
        if a % i == 0:
            tong += i
    
    if tong == a:
        return True
    else:
        return False
    
x = int(input())
print(sohoanhao(x))
```

## W5A5: Tìm vị trí (index) đầu tiên của k trong list

```python
def Find_Index(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i 
            
    return -1

```

## W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 

```python
def giaithua(n):
    ketqua = 1
    for i in range(1, n+1):
        ketqua = ketqua * i
        
    return ketqua
    
b = int(input())
print(giaithua(b))
```

## W5A7 


```python
def Calculator(num1, num2, operat):
    if operat == '+':
        res = num1 + num2
    elif operat == '-':
        res = num1 - num2
    elif operat == '*':
        res = num1 * num2
    elif operat == '/':
        if num2 == 0:
            return "Error: Division by zero"
        res = num1 / num2
    else:
        return "Invalid operator"
    

    return round(res, 2)
```

## W5A9: Tổng các chữ số của n

```python
def tong(n):
    n = abs(n)
    tong = 0
    while n > 0:
        tong += n % 10
        n = n // 10
    return(tong)

a = int(input())
print(tong(a))
```

## W5A11 Viết hàm is_product_divisible_by_10(a, b) trả về True nếu tích a*b chia hết cho 10, ngược lại False

```python
def chiahet(a,b):
    if (a*b % 10 == 0):
        return True
    else:
        return False


x, y = map(int, input().split())
print(chiahet(x,y))
```
