## Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input("Nhập độ cao h: "))

for i in range(1, h + 1):
    print(" " * (h - i), end="")
    
    if i == h:
        print("*" * (2 * i - 1))
    elif i == 1:
        print("*")
    else:
        print("*", end="")
        print(" " * (2 * i - 3), end="")
        print("*")
```

## Bài 5: Nhập vào số nguyên dương n. Tính S = 1 + 2 + 3 + 4 + … + n

```python
n = int(input("Nhập số nguyên dương n: "))
S = 0
for i in range(1, n + 1):
    S += i
print(f"Tổng S = {S}")
```

## Bài 8: In toàn bộ ước chung của a và b

## Bài 10: Kiểm tra số hoàn hảo

```python
a = int(input("Nhập a: "))
tong_uoc = 0
for i in range(1, a):
    if a % i == 0:
        tong_uoc += i

if tong_uoc == a:
    print(f"{a} là số hoàn hảo")
else:
    print(f"{a} không phải là số hoàn hảo")
```

## Bài 12: Tìm k sao cho S(k) < n lớn nhất

```python
n = int(input("Nhập số nguyên dương n: "))
S = 0
k = 0
while S + (k + 1) < n:
    k += 1
    S += k

print(f"Giá trị k tìm được là: {k}")
print(f"Tổng S({k}) tương ứng là: {S}")
```

## Bài 13: Tìm n nhỏ nhất sao cho $1 + 1/2 + ... + 1/n > A$

```python
A = float(input("Nhập số nguyên dương A: "))
S = 0
n = 0
while S <= A:
    n += 1
    S += 1/n
print(f"n nhỏ nhất là: {n}")
```

## Bài 14: Tìm Max, Min của dãy số (dừng khi nhập -1, không dùng mảng)

```python
max_val = -float('inf') 
min_val = float('inf')  
da_nhap_so = False

while True:
    num = int(input("Nhập số (nhập -1 để dừng): "))
    if num == -1:
        break
    
    da_nhap_so = True
    if num > max_val: max_val = num
    if num < min_val: min_val = num

if da_nhap_so:
    print(f"Lớn nhất: {max_val}, Nhỏ nhất: {min_val}")
else:
    print("Bạn chưa nhập số nào.")
```

## Bài 15: Đếm số chữ số của n (Nhiều cách nhất có thể)

```python
n = int(input("Nhập n: "))
count = 0
temp = abs(n)
if temp == 0: count = 1
while temp > 0:
    count += 1
    temp //= 10
print(f"Cách 1: {count}")
```

## Bài 16: Đếm chữ số chẵn, chữ số lẻ

```python
n = abs(int(input("Nhập n: ")))
chan = 0
le = 0
if n == 0: 
    chan = 1
while n > 0:
    chu_so = n % 10
    if chu_so % 2 == 0: 
        chan += 1
    else: 
        le += 1
    n //= 10
print(f"Chẵn: {chan}, Lẻ: {le}")
```

## Bài 17: Tính tổng các chữ số của n

```python
n = abs(int(input("Nhập n: ")))
tong = 0
while n > 0:
    tong += n % 10
    n //= 10
print(f"Tổng các chữ số: {tong}")
```

## Bài 18: Kiểm tra n có dạng $3^k$ hay không

```python
n = int(input("Nhập n: "))
if n <= 0:
    print(False)
else:
    temp = n
    while temp % 3 == 0:
        temp //= 3
    if temp == 1:
        print(f"{n} có dạng 3^k")
    else:
        print(f"{n} không có dạng 3^k")
```

## Bài 19: Số Fibonacci lớn nhất không vượt quá A

```python
A = int(input("Nhập A: "))
f1, f2 = 1, 1
while f1 + f2 <= A:
    next_f = f1 + f2
    f1 = f2
    f2 = next_f


print(f"Số Fibonacci lớn nhất không quá A là: {f2 if A >= 1 else 'Không có'}")
```

## Bài 20: Đếm chữ số của n là số nguyên tố (2, 3, 5, 7)

```python
n = abs(int(input("Nhập n: ")))
count = 0
while n > 0:
    dv = n % 10
    if dv == 2 or dv == 3 or dv == 5 or dv == 7:
        count += 1
    n //= 10
print(f"Số chữ số là số nguyên tố: {count}")
```
