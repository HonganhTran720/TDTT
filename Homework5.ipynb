## W5A16 
Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 

```python
def is_armstrong(x):
    # Số âm không phải là số Armstrong
    if x < 0:
        return False
    
    # Bước 1: Tìm số chữ số (n)
    n = 0
    temp = x
    if temp == 0:
        n = 1
    else:
        while temp > 0:
            temp //= 10  # Chia lấy nguyên để bỏ chữ số cuối
            n += 1
            
    # Bước 2: Tính tổng lũy thừa bậc n của các chữ số
    tong = 0
    temp = x
    while temp > 0:
        chu_so = temp % 10   # Lấy chữ số cuối cùng
        tong += chu_so ** n   # Cộng lũy thừa vào tổng
        temp //= 10          # Bỏ chữ số cuối đã tính
        
    # Bước 3: Kiểm tra và trả về kiểu bool
    return tong == x

# Chạy thử
n = int(input("Nhập số nguyên x: "))
print(is_armstrong(n))
```

## W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên. 


```python
def sodoixung(a):
    if a<0:
        return False
    
    goc = a
    sodaonguoc = 0
    while a > 0:
        socuoi = a % 10
        sodaonguoc = a * 10 + socuoi
        a //= 10
    return sodaonguoc == goc

n = int(input())
print(sodoixung(n))

```

## W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def sokydieu(n):
    temp = n
    while temp > 9:
        tong = 0
        while temp > 0:
            tong += temp % 10
            temp //= 10
        
        temp = tong
    return temp == 1

a = int(input())
print(sokydieu(a))
```

## W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except ValueError:
        pass

    try:
        float(s)
        return 'float'
    except ValueError:
        return 'neither'
    
a = input()
print(detect_number_type(a))
```

## W5A21 Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại 

```python
def is_one_product_of_others(a, b, c):
    return (a == b * c) or (b == a * c) or (c == a * b)
```

## W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không. 

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    return (y2 - y1) * (x3 - x2) == (y3 - y2) * (x2 - x1)

```

## W5A25 Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay  không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), (0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    def cross_product(a, b, c):
        return (b[0] - a[0]) * (c[1] - a[1]) - (b[1] - a[1]) * (c[0] - a[0])

    cp = []
    for i in range(4):
        p1 = quad[i]
        p2 = quad[(i + 1) % 4]
        cp.append(cross_product(p1, p2, p))
    
    return all(x >= 0 for x in cp) or all(x <= 0 for x in cp)
```

## W5A27: Đếm chữ số khác 0 trong danh sách số nguyên

```python
def sokhackhong(numbers):
    dem = 0
    for n in numbers:
        # Chuyển số thành chuỗi, bỏ dấu '-' nếu có và đếm ký tự khác '0'
        s = str(abs(n))
        dem += sum(1 for digit in s if digit != '0')
    return dem
```

## W5A30: Tính hệ số nhị thức $C(n, k)$

```python
def combination(n, k):
    if k < 0 or k > n:
        return 0
    if k == 0 or k == n:
        return 1

    if k > n // 2:
        k = n - k
        
    res = 1
    for i in range(k):
        res = res * (n - i) // (i + 1)
        
    return res

a, b = map(int, input().split())
print(combination(a, b))
```

## W5A31: Tổng bình phương từ 1 đến n

```python
def sum_squares(n):
    return sum(i**2 for i in range(1, n + 1))
```

## W5A34: Tổng các số có chữ số đôi một khác nhau

```python
def has_unique_digits(x):
    s = str(abs(x))
    return len(set(s)) == len(s)

def sum_1_to_n(n):
    if n <= 0:
        return 0
    total = 0
    for x in range(1, n + 1):
        if has_unique_digits(x):
            total += x
    return total
```
